```python
def convert_scores(score_key):
    image_scores = {}
    for text, score in score_key.items():
        doc = nlp(text)
        for token in doc:
            if(token.pos_ == "NOUN"):
                image_scores[token.lemma_] = score 
    return image_scores
```

```python

def update_scores(row):
    scores = row.scores
    p_scores = row.processed
    for text, score in scores.items():
        tokens = [val.strip().lower() for val in text.split(",")]
        for token in tokens:
            if token not in p_scores:
                p_scores[token] = score
    row.processed = p_scores
    return row
update_scores(output.iloc[4])
```
