```python
import gensim
```

```python
import pandas as pd
from gensim.models.ldamodel import LdaModel
```

# Loading documents 

```python
data = pd.read_csv("extracted.csv")
data.head()
```

```python
documents = data.caption.values
documents
```

# Removing stop words

```python
stoplist = set('for a of the and to in on is'.split())
```

```python
texts = [[word for word in document.lower().split() if word not in stoplist] for document in documents]
texts
```

# Dictionary

```python
dictionary = gensim.corpora.Dictionary(texts)
```

```python
print(dictionary)
```

```python
print(dictionary.token2id)
```

```python
corpus = [dictionary.doc2bow(text) for text in texts]
```

```python
corpus[:10]
```

```python
lda = LdaModel(corpus, num_topics=25, id2word = dictionary, passes=50)
```

```python
lda.show_topics(num_topics=25, num_words=10)
```
